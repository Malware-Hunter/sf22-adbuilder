```python
import glob

time_geration = []
time_convert = []
time_concat = []

for file in glob.glob('*.log'):
    print("\n*** abrindo arquivo:", file)
    if "Concat" in file:
        with open(file, 'r') as f:
            for line in f:
                #print(line)
                if "INT" in line:
                    try:
                        split = str(line).split('  ') # split[1] - tempo de execução (segundos)
                        time_convert.append(float(split[1]))
                    except:
                        ''
                if "CSV" in line:
                    try:
                        split = str(line).split('  ') # split[1] - tempo de execução (segundos)
                        split_aux = str(split[1]).split(' ')
                        time_concat.append(float(split_aux[0]))
                    except:
                        ''
                
    elif "Geration" in file:
        with open(file, 'r') as f:
            for line in f:
                if "Tempo" in line:
                    try:
                        split = str(line).split(' ') # split[7] - tempo de execução (segundos)
                        time_geration.append(float(split[7]))
                    except:
                        ''
```

```python
import pandas as pd

df = pd.DataFrame({'Tempo de Geração (s)': time_geration, 'Tempo de Conversão para INT (s)': time_convert, 'Tempo de Concatenação (s)': time_concat})
df['Tempo de building (s)'] = df['Tempo de Geração (s)'] + df['Tempo de Conversão para INT (s)'] + df['Tempo de Concatenação (s)']
df.head()
```
